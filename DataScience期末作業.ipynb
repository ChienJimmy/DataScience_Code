# 資料處理

* 107學年度冷熱門校系學生人數以及整體占比

```python
import pandas as pd
import plotly.express as px
df07 = pd.read_csv("https://stats.moe.gov.tw/files/detail/107/107_sdata.csv")

# 用0取代-(無資料即視為0位學生)
df07["學生數"].replace({"-": 0}, inplace=True)
# 將"1,025"(str)轉換為1025(int)
for i in range(len(df07["學生數"])):
  strs = str(df07.loc[i, "學生數"]).replace(",", "")
  df07.loc[i, "學生數"] = int(strs)
# 擷取台大的資料，並去除0位學生之科系
NTU07 = df07[((df07["學校代碼"] == 3) & (df07['學生數'] != 0))].loc[:, ["科系名稱", "學生數"]]
# 將同校系名稱的資料合併
group07 = NTU07.groupby("科系名稱")
info07 = group07.sum()
# 計算比例
stu07 = df07.loc[:,["學校代碼", "學生數"]].groupby("學校代碼").sum().iloc[2,0]
info07.insert(1, column="107學年度學生數占比(%)", value=(100 * (info07["學生數"] / stu07)))
# 改index
info07 = info07.rename(columns={"學生數":"107學年度學生數"})
# 繪圖前處理
info07 = info07.reset_index()
info07_hotest = info07.nlargest(30, "107學年度學生數")
info07_coldest = info07.nsmallest(30, "107學年度學生數").sort_values(by="107學年度學生數",ascending=False)
# 製圖
num07 = px.bar(pd.concat([info07_hotest, info07_coldest]), x="科系名稱", y="107學年度學生數", width=1600, height=800)
num07.show()
per07 = px.bar(pd.concat([info07_hotest, info07_coldest]), x="科系名稱", y="107學年度學生數占比(%)", width=1600, height=800)
per07.show()
```

* 台大前15熱門系所近五年學生數變化綜覽

```python
import pandas as pd
import plotly.express as px
df_07 = pd.read_csv("https://stats.moe.gov.tw/files/detail/107/107_sdata.csv")
df_08 = pd.read_csv("https://stats.moe.gov.tw/files/detail/108/108_sdata.csv")
df_09 = pd.read_csv("https://stats.moe.gov.tw/files/detail/109/109_sdata.csv")
df_10 = pd.read_csv("https://stats.moe.gov.tw/files/detail/110/110_sdata.csv")
df_11 = pd.read_csv("https://stats.moe.gov.tw/files/detail/111/111_sdata.csv")
# 用0取代-(無資料即視為0位學生)
for df in [df_07, df_08, df_09, df_10, df_11]:
  df["學生數"].replace({"-": 0}, inplace=True)

# 將"1,025"(str)轉換為1025(int)
for df in [df_07, df_08, df_09, df_10, df_11]:
  for i in range(len(df["學生數"])):
    strs = str(df.loc[i, "學生數"]).replace(",", "")
    df.loc[i, "學生數"] = int(strs)

# 擷取台大的資料，並去除0位學生之科系
NTU_07 = df_07[((df_07["學校代碼"] == 3) & (df_07['學生數'] != 0))].loc[:,["科系名稱", "學生數"]]
NTU_08 = df_08[((df_08["學校代碼"] == 3) & (df_08['學生數'] != 0))].loc[:,["科系名稱", "學生數"]]
NTU_09 = df_09[((df_09["學校代碼"] == 3) & (df_09['學生數'] != 0))].loc[:,["科系名稱", "學生數"]]
NTU_10 = df_10[((df_10["學校代碼"] == 3) & (df_10['學生數'] != 0))].loc[:,["科系名稱", "學生數"]]
NTU_11 = df_11[((df_11["學校代碼"] == 3) & (df_11['學生數'] != 0))].loc[:,["科系名稱", "學生數"]]

# 將同校系名稱的資料合併後，將index的學生數改名為該學年度學生數，並取學生數前15名之系所
info_07 = NTU_07.groupby("科系名稱").sum()
info_07 = info_07.rename(columns={"學生數":"107學年度學生數"})
info_07 = info_07.nlargest(15, "107學年度學生數")
info_08 = NTU_08.groupby("科系名稱").sum()
info_08 = info_08.rename(columns={"學生數":"108學年度學生數"})
info_08 = info_08.nlargest(15, "108學年度學生數")
info_09 = NTU_09.groupby("科系名稱").sum()
info_09 = info_09.rename(columns={"學生數":"109學年度學生數"})
info_09 = info_09.nlargest(15, "109學年度學生數")
info_10 = NTU_10.groupby("科系名稱").sum()
info_10 = info_10.rename(columns={"學生數":"110學年度學生數"})
info_10 = info_10.nlargest(15, "110學年度學生數")
info_11 = NTU_11.groupby("科系名稱").sum()
info_11 = info_11.rename(columns={"學生數":"111學年度學生數"})
info_11 = info_11.nlargest(15, "111學年度學生數")

# 用pd.merge合併資料
Merged = pd.merge(info_07, info_08, on=['科系名稱'])
Merged = pd.merge(Merged, info_09, on=['科系名稱'])
Merged = pd.merge(Merged, info_10, on=['科系名稱'])
Merged = pd.merge(Merged, info_11, on=['科系名稱'])

# 繪製折線圖
fig = px.line(Merged.T, y=['法律學系', '機械工程學系', '電機工程學系', '政治學系', '醫學系', '土木工程學系', '資訊工程學系', '經濟學系', '化學工程學系', '財務金融學系', '會計學系', '外國語文學系', '化學系', '國際企業學系', '物理學系'],width=800, height=500)
fig.show()
```
